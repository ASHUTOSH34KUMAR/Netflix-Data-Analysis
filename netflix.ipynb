# Convert to Date Time Format

```python
df['date_added']=df['date_added'].astype('string').str.strip()
df["date_added"]=pd.to_datetime(df['date_added'])
df['day_added']=df['date_added'].dt.day
df['year_added']=df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
```

# Data Visualization

Type : Movie and TV shows

```python
count = df['type'].value_counts()
count
```

```python
plt.bar(['Movie','Tv Show'],count)
```

```python
sns.countplot(netflix['type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

```python
x=netflix['rating'].value_counts()
plt.figure(figsize=(12,6))
colors = plt.cm.tab20.colors[:len(x)]
plt.bar(x.index,x.values,color=colors)
plt.xticks(rotation=90,ha='right')
plt.xlabel('Rating',fontsize=14,fontweight='bold')
plt.ylabel('No.of Occurances',fontsize=14,fontweight='bold')
plt.show()
```

```python
x = netflix.groupby('type')['rating'].value_counts().unstack().fillna(0)
y1=list(x.loc['Movie',:])
y2=list(x.loc['TV Show',:])
x1 = list(x.columns)
z1 = list(x.index)
barwidth = 0.33
x = np.arange(0,len(x1),1)
x2 = [y + barwidth for y in x]
plt.figure(figsize=(12,6))
plt.xticks([x1 + barwidth/2 for x1 in x],x1)
plt.bar(x,y1,width=barwidth)
plt.bar(x2,y2,width=barwidth)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Relation between Rating and count')
plt.legend(z1)
```

```python
labels = ['Movie','Tv show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0,1,2))
plt.figure(figsize=(9,9))
plt.pie(size,labels=labels,colors=colors,explode=[0,0.1],shadow=True,startangle=90)
plt.title('DIstribution of Type',fontsize=25)
plt.legend()
plt.show()
```

```python
x = netflix['rating'].value_counts()
plt.figure(figsize=(10,8))
plt.pie(x,labels=list(x.index),autopct='%1.1f%%',shadow=True)
plt.show()
```
